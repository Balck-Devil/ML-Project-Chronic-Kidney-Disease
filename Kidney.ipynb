```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
ind_col = [col for col in data.columns if col!="class"]
dep_col = "class"
```

```python
X = data[ind_col]
y = data[dep_col]
```

```python
ordered_rank_feature = SelectKBest(score_func=chi2,k=20)
ordered_features = ordered_rank_feature.fit(X,y)
ordered_features
```

```python
ordered_features.scores_
```

```python
datascores = pd.DataFrame(ordered_features.scores_,columns=["Score"])
datascores
```

```python
dfcolumns = pd.DataFrame(X.columns)
dfcolumns
```

```python
features_rank = pd.concat([dfcolumns,datascores],axis=1)
features_rank
```

```python
features_rank.columns=['Features','Score']
features_rank
```

```python
features_rank.nlargest(10,"Score")
```
